 **Problem Description**
Kakuma Refugee Camp continues to experience persistent economic exclusion, particularly among
youth and women who face limited access to formal employment, financial capital, and skills
development opportunities. Despite ongoing NGO interventions, structural barriers such as limited
education pathways, restricted movement, and underdeveloped local markets prevent meaningful
income generation. High youth unemployment contributes to dependency, poverty, and increased
social vulnerability. A data-driven analysis of unemployment patterns is essential to inform targeted,
scalable livelihood interventions that promote self-reliance and inclusive economic growth.


**Objectives**


1.Identify unemployment rates across key demographic groups (age, gender, education level)


2. Analyze the relationship between education/vocational training and income levels


3.Assess access to NGO livelihood programs and financial services


4. Identify structural barriers limiting access to employment opportunities



5. Generate actionable, data-driven recommendations for stakeholder

**Expolatory Data Analysis**

**Loading  Inspecting and Cleaning Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

sns.set_theme(style="whitegrid")

# Mount Google Drive
drive.mount('/content/drive')
Mount_Drive = '/content/drive'

# Load dataset
df = pd.read_csv(f'{Mount_Drive}/MyDrive/kakuma_youth_unemployment (1).csv')

# Clean column names
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

# Drop missing values
df = df.dropna()

# Convert types
df['age'] = df['age'].astype(int)
df['monthly_income_kes'] = df['monthly_income_kes'].astype(float)
```

**Objective 2**;Analyze the relationship between education/vocational training and income levels

**Business Question**:What is average income by education level?

```python
income_edu = df.groupby('education_level')['monthly_income_kes'].mean()
income_edu.plot(kind='bar')
plt.title("Income by Education Level")
plt.show()
```

**Business Question**: Does vocational training increase income?

```python
income_training = df.groupby('vocational_training_completed')['monthly_income_kes'].mean()

income_training.plot(kind='bar')
plt.title("Income: Training vs No Training")
plt.show()
```

**Business Question**:  What is a combined effect of education and training?

```python
pivot = df.pivot_table(values='monthly_income_kes',
                       index='education_level',
                       columns='vocational_training_completed',
                       aggfunc='mean')

pivot.plot(kind='bar', figsize=(10, 6))
plt.title("Income by Education & Training")
plt.ylabel("Average Monthly Income (KES)")
plt.xlabel("Education Level")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Vocational Training')
plt.tight_layout()
plt.show()
```

**Objective 3**: Assess access to NGO livelihood programs and financial services

**Business Question**: What % of youth have NGO support

```python
ngo_access = df['ngo_livelihood_program_participation'].value_counts(normalize=True)*100
ngo_access.plot(kind='pie', autopct='%1.1f%%')
plt.title("NGO Access Distribution")
plt.show()
```

**Business Question**: Do NGO programs improve income?

```python
ngo_income = df.groupby('ngo_livelihood_program_participation')['monthly_income_kes'].mean()

ngo_income.plot(kind='barh', color='orange')
plt.title("Income by NGO Support")
plt.show()
```

**Business Question**: Do NGO programs improve employment?

```python
ngo_emp = df.groupby('ngo_livelihood_program_participation')['employment_status'].apply(lambda x: (x=='Employed').mean()*100)

ngo_emp.plot(kind='bar')
plt.title("Employment Rate by NGO Access")
plt.show()
```

**Business Question**:What % lack financial access?

```python
fin_access = df['access_to_microfinance'].value_counts(normalize=True)*100

fin_access.plot(kind='pie', autopct='%1.1f%%')
plt.title("Financial Access Distribution (Microfinance)")
plt.show()
```

**Objective 4:Identify structural barriers limiting access to employment opportunities**

**Business Question**:What are the top employment barriers?

```python
barriers = df['primary_employment_barrier'].value_counts()

barriers.plot(kind='barh', color='brown')
plt.title("Top Employment Barriers")
plt.show()
```

**Business Question**; Which barriers affect women most?

```python
barriers_gender = df.groupby(['gender','primary_employment_barrier']).size().unstack()

barriers_gender.plot(kind='bar', stacked=True)
plt.title("Barriers by Gender")
plt.show()
```

Business Question: How do barriers affect income?

```python
barrier_income = df.groupby('primary_employment_barrier')['monthly_income_kes'].mean()

barrier_income.plot(kind='barh', color='grey')
plt.title("Income by Barrier Type")
plt.show()
```
